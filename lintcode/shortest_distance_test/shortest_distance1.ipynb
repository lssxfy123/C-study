# 时间复杂度为O(houseCount * n * m)，其中houseCount为房子的数量，最坏情况下为O(n^2 * m^2)
# 代码相比之前不超时的原因是houseCount可能远小于emptyCount，它们的时间复杂度其实是相同的

```python
class GridType:
    EMPTY = 0
    HOUSE = 1
    WALL = 2
```

```python
class Solution:
    """
    @param grid: a 2D grid
    @return: An integer
    """
    def shortestDistance(self, grid):
        # write your code here
        if not grid or not grid[0]:
            return -1
        n, m = len(grid), len(grid[0])
        positions = [(i, j) for j in range(m) for i in range(n) if grid[i][j] == GridType.EMPTY]
        homes = [(i, j) for j in range(m) for i in range(n) if grid[i][j] == GridType.HOUSE]
        home_count = len(homes)
        
        ## key为空地坐标，value为到所有能到达的房子的距离
        distance_sum = {}
        
        ## key为空地坐标，value为能到达的房子数
        reachable_count = {}
        
        min_distance = float('inf')
        ## 每个可以做邮局的位置，都作为起点进行bfs搜索
        ## distance存储的是每个空地和房子到you邮局的距离
        for i, j in homes:
            self.get_distance_by_bfs(i, j, grid, distance_sum, reachable_count)
        
        for i in range(n):
            for j in range(m):
                if (i, j) not in distance_sum:
                    continue
                
                if reachable_count[(i, j)] != home_count:
                    continue
                min_distance = min(min_distance, distance_sum[(i, j)])
            
        if min_distance != float('inf'):
            return min_distance
        return -1
            
    
    def get_distance_by_bfs(self, i, j, grid, distance_sum, reachable_count):
        ## 第一个key(i, j)为某个房子的坐标，其它key为该房子能到达的空地坐标
        ## value为空地到(i, j)的距离
        distance = {(i, j) : 0}
        queue = collections.deque([(i, j)])
        directions = [(0, 1), (0, -1), (1, 0), (-1, 0)]
        while queue:
            x, y = queue.popleft()
            for dx, dy in directions:
                new_x, new_y = x + dx, y + dy
                if not self.is_valid(new_x, new_y, grid):
                    continue
                if (new_x, new_y) not in distance:
                    distance[(new_x, new_y)] = distance[(x, y)] + 1
                    queue.append((new_x, new_y))
                    
                    if (new_x, new_y) not in distance_sum:
                        distance_sum[(new_x, new_y)] = 0
                        reachable_count[(new_x, new_y)] = 0
                    distance_sum[(new_x, new_y)] += distance[(new_x, new_y)]
                    reachable_count[(new_x, new_y)] += 1

    
    def is_valid(self, x, y, grid):
        n, m = len(grid), len(grid[0])
        if 0 <= x < n and 0 <= y < m and grid[x][y] == GridType.EMPTY:
            return True
        return False
            
```
