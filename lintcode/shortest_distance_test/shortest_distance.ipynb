# 时间复杂度为O(emptyCount*n*m)，其中emptyCount为空地的数量，最坏情况下为O(n^2*m^2)
# 下面的代码会超时，推测是由于测试样例中emptyCount非常多，导致超时

```python
class Solution:
    """
    @param grid: a 2D grid
    @return: An integer
    """
    def shortestDistance(self, grid):
        # write your code here
        if not grid or not grid[0]:
            return -1
        n, m = len(grid), len(grid[0])
        positions = [(i, j) for j in range(m) for i in range(n) if grid[i][j] == 0]
        homes = [(i, j) for j in range(m) for i in range(n) if grid[i][j] == 1]
        home_count = len(homes)
        
        min_distance = float('inf')
        ## 每个可以做邮局的位置，都作为起点进行bfs搜索
        ## distance存储的是每个空地和房子到you邮局的距离
        for i, j in positions:
            min_distance = min(self.get_distance_by_bfs(i, j, grid, home_count), min_distance)
            
        if min_distance != float('inf'):
            return min_distance
        return -1
            
    
    def get_distance_by_bfs(self, i, j, grid, home_count):
        distance = {(i, j) : 0}
        queue = collections.deque([(i, j)])
        directions = [(0, 1), (0, -1), (1, 0), (-1, 0)]
        distance_sum = 0
        count = 0
        while queue:
            x, y = queue.popleft()
            for dx, dy in directions:
                new_x, new_y = x + dx, y + dy
                if not self.is_valid(new_x, new_y, grid):
                    continue
                if (new_x, new_y) in distance:
                    continue
                distance[(new_x, new_y)] = distance[(x, y)] + 1
                ## 如果是房子，就停止不再进行遍历
                if grid[new_x][new_y] == 1:
                    count += 1
                    distance_sum += distance[(new_x, new_y)]
                    continue
                queue.append((new_x, new_y))
        if count != home_count:
            return float('inf')
        return distance_sum
    
    def is_valid(self, x, y, grid):
        n, m = len(grid), len(grid[0])
        if 0 <= x < n and 0 <= y < m and grid[x][y] != 2:
            return True
        return False
            
```
