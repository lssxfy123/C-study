```python
## 暴力做法，先求出树的中序遍历，然后使用之前双指针中的最接近的k个数
class Solution:
    """
    @param root: the given BST
    @param target: the given target
    @param k: the given k
    @return: k values in the BST that are closest to the target
    """
    def closestKValues(self, root, target, k):
        # write your code here
        results = []
        inorder_result = self.inorder_traversal(root)
        
        ## 查找左边小于target，右边大于等于target的第1个位置
        upper_index = self.find_upper_index(inorder_result, target)
        left = upper_index - 1
        right = upper_index
        
        for _ in range(k):
            if self.is_left_closer(left, right, inorder_result, target):
                results.append(inorder_result[left])
                left -= 1
            else:
                results.append(inorder_result[right])
                right += 1
        
        return results
    
    ## 中序遍历
    def inorder_traversal(self, root):
        traversal = []
        stack = []
        while root:
            stack.append(root)
            root = root.left
        
        while stack:
            node = stack[-1]
            traversal.append(node.val)
            if node.right:
                node = node.right
                while node:
                    stack.append(node)
                    node = node.left
            else:
                node = stack.pop()
                while stack and stack[-1].right == node:
                    node = stack.pop()
        return traversal
    
    ## 查找第1个大于等于target的位置
    def find_upper_index(self, nums, target):
        left, right = 0, len(nums) - 1
        
        ## 二分法要注意判断条件left + 1 < right
        ## 否则容易死循环
        while left + 1 < right:
            mid = (left + right) // 2
            if nums[mid] >= target:
                right = mid
            else:
                left = mid
        
        if nums[left] >= target:
            return left
        
        if nums[right] >= target:
            return right
        return len(nums)
    
    def is_left_closer(self, left, right, nums, target):
        if left < 0:
            return False
        
        if right >= len(nums):
            return True
        
        return target - nums[left] < nums[right] - target
        
```
