```python
## 归并排序
class Solution:
    """
    @param A: an integer array
    @return: nothing
    """
    def sortIntegers2(self, A):
        # write your code here
        if not A or len(A) <= 1:
            return
        
        ## 归并排序需要开辟额外的空间n，空间复杂度为O(n)
        result = [0 for _ in range(len(A))]
        self.merge_sort(A, 0, len(A) - 1, result)
    
    
    def merge_sort(self, A, start, end, result):
        if start >= end:
            return
        
        mid = (start + end) // 2
        self.merge_sort(A, start, mid, result)
        self.merge_sort(A, mid + 1, end, result)
        
        ## 归并
        self.merge(A, start, mid, end, result)
        
    def merge(self, A, start, mid, end, result):
        index_first = start
        index_second = mid + 1
        index = start
        while index_first <= mid and index_second <= end:
            if A[index_first] < A[index_second]:
                result[index] = A[index_first]
                index_first += 1
            else:
                result[index] = A[index_second]
                index_second += 1
            index += 1
        
        while index_first <= mid:
            result[index] = A[index_first]
            index_first += 1
            index += 1
        
        while index_second <= end:
            result[index] = A[index_second]
            index_second += 1
            index += 1
        
        ## 归并的最后要修改数组A本身
        index = start
        while index <= end:
            A[index] = result[index]
            index += 1
            
```

```python
solution = Solution()
A = [19,-10,-2,40,3,36,57,25,66,51,5,40,-8,43,9,-5,0,4,55,28,63,67,-2,35,57,0,0,30,17,55,22,34,-4,42,57,52,4,65,6,-2,8,12,31,43,26,34,31,19,-3,36,34,11,58,23,13,7,17,10,33,-5,10,53,14,56,18,8,-6,-2,37,7]
# A = [-6, -5, -5]
solution.sortIntegers2(A)
print(A)
```
