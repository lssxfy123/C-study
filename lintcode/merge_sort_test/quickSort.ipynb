```python
## 快速排序
class Solution:
    """
    @param A: an integer array
    @return: nothing
    """
    def sortIntegers2(self, A):
        # write your code here
        if not A or len(A) <= 1:
            return
        self.quick_sort(A, 0, len(A) - 1)
    
    def quick_sort(self, A, start, end):
        if start >= end:
            return
        
        left = start
        right = end
        ## 1. 枢纽元的选择：枢纽元一般选择中间值，另外也可以随机选或通过3中位法进行选择
        pivot = A[(start + end) // 2]
        
        ## 2. 左右指针的比较要包含等于的情况，主要是为了后序分割数组的范围考虑，要确保while
        ## 循环结束后，left>right，这样递归调用时，start到right的区间和left到end的区间不会重叠
        ## 快速排序中，分割后的两个区间一定不能重叠，否则会报递归层次太深的错误
        while left <= right:
            ## 3. 与枢纽元进行比较时，左右指针指向的元素都不要包含等于pivot的情况
            ## 这是考虑极端情况，例如数组是[1, 1, 1, 1]，枢纽元肯定是1，如果在比较时
            ## 包含等于的情况，A[left]<=pivot或A[right]>=pivot，会出现left或right越界的情况
            ## 即使进行保护，确保left或right不越界，也会出现把数组分割成一个空数组和一个包含
            ## 全部数组的情况，或者一个很小的数组和一个很大的数组的情况，这样快速排序的时间复杂度
            ## 就会退步成最坏的情况
            while left <= right and A[left] < pivot:
                left += 1
                
            while left <= right and A[right] > pivot:
                right -= 1
            
            if left <= right:
                A[left], A[right] = A[right], A[left]
                left += 1
                right -= 1

        self.quick_sort(A, start,right)
        self.quick_sort(A, left, end)
            
```

```python
solution = Solution()
A = [19,-10,-2,40,3,36,57,25,66,51,5,40,-8,43,9,-5,0,4,55,28,63,67,-2,35,57,0,0,30,17,55,22,34,-4,42,57,52,4,65,6,-2,8,12,31,43,26,34,31,19,-3,36,34,11,58,23,13,7,17,10,33,-5,10,53,14,56,18,8,-6,-2,37,7]
# A = [-6, -5, -5]
solution.sortIntegers2(A)
print(A)
```
